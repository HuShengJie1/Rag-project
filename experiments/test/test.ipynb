```python
"""
Utility function for showing images.

Intended to be imported in Jupyter notebooks to display pixmap images.

Invocation: "show_image(item, title)", where item is a PyMuPDF object
which has a "get_pixmap" method, and title is an optional string.

The function executes "item.get_pixmap(dpi=150)" and show the resulting
image.


Dependencies
------------
numpy, matplotlib, pymupdf
"""


def show_image(item, title=""):
    """Display a pixmap.

    Just to display Pixmap image of "item" - ignore the man behind the curtain.

    Args:
        item: any PyMuPDF object having a "get_pixmap" method.
        title: a string to be used as image title

    Generates an RGB Pixmap from item using a constant DPI and using matplotlib
    to show it inline of the notebook.
    """
    DPI = 150  # use this resolution
    import numpy as np
    import matplotlib.pyplot as plt

    # %matplotlib inline
    pix = item.get_pixmap(dpi=DPI)
    img = np.ndarray([pix.h, pix.w, 3], dtype=np.uint8, buffer=pix.samples_mv)
    plt.figure(dpi=DPI)  # set the figure's DPI
    plt.title(title)  # set title of image
    _ = plt.imshow(img, extent=(0, pix.w * 72 / DPI, pix.h * 72 / DPI, 0))
```

```python
import fitz  # PyMuPDF
from pathlib import Path
from marker.converters.pdf import PdfConverter
from marker.models import create_model_dict
from marker.output import text_from_rendered

def test_marker_official_by_page(input_pdf: str, output_md: str):
    """
    使用官方最新 API 逐页处理 PDF，并注入页码标记。
    """
    input_path = Path(input_pdf)
    if not input_path.exists():
        print(f"❌ 找不到文件: {input_pdf}")
        return

    # 1. 初始化转换器（只需加载一次，对应你提供的官方写法）
    print("⏳ 正在初始化 Marker 模型字典（此过程较慢，请稍候）...")
    converter = PdfConverter(
        artifact_dict=create_model_dict(),
    )
    
    # 2. 打开原始 PDF 并创建临时目录
    doc = fitz.open(input_pdf)
    total_pages = len(doc)
    temp_dir = Path("./temp_split_pages")
    temp_dir.mkdir(exist_ok=True)
    
    all_page_contents = []

    try:
        for i in range(total_pages):
            current_page_num = i + 1
            print(f"🔄 [进度 {current_page_num}/{total_pages}] 正在处理中...")

            # 3. 提取单页并保存为临时文件
            temp_pdf_path = temp_dir / f"page_{current_page_num}.pdf"
            single_page_doc = fitz.open()
            single_page_doc.insert_pdf(doc, from_page=i, to_page=i)
            single_page_doc.save(str(temp_pdf_path))
            single_page_doc.close()

            # 4. 调用官方最新接口进行转换
            # 使用你提供的官方 API 流程
            rendered = converter(str(temp_pdf_path))
            text, _, _ = text_from_rendered(rendered)

            # 5. 注入自定义页码标记 (HTML 注释格式)
            # 这种格式在 Markdown 渲染中是隐藏的，但对我们的分块脚本（Chunking）可见
            marked_text = f"\n\n\n{text}\n\n"
            all_page_contents.append(marked_text)

            # 清理单页临时文件
            temp_pdf_path.unlink()

        # 6. 合并所有页面内容并保存
        final_md_path = Path(output_md)
        with open(final_md_path, "w", encoding="utf-8") as f:
            f.write("\n---\n".join(all_page_contents))
        
        print(f"\n✨ 处理完成！最终文件：{final_md_path.absolute()}")

    finally:
        # 彻底清理临时目录
        if temp_dir.exists():
            for f in temp_dir.glob("*"): f.unlink()
            temp_dir.rmdir()
        doc.close()

if __name__ == "__main__":
    # --- 测试配置 ---
    # 建议先找一个页数较少的 PDF（如 2-3 页）进行测试
    test_pdf = "附件1-5  上海海洋大学信息学院课程体系合理性评价办法.pdf" 
    output_result = "marker_test_official.md"
    
    test_marker_official_by_page(test_pdf, output_result)
```

```python
"""基于 LangChain 的增强型分块模块。

支持从带有 ==== PAGE_X_START ==== 标记的 Markdown 中提取物理页码元数据。
"""

from __future__ import annotations

import re
from pathlib import Path
from typing import Dict, List, Sequence

from langchain_text_splitters import MarkdownHeaderTextSplitter, RecursiveCharacterTextSplitter

__all__ = ["chunk_markdown_file", "chunk_all_documents"]

Chunk = Dict[str, object]

def _build_chunk(
    source_file: Path,
    text: str,
    seq: int,
    metadata: Dict[str, object] | None = None,
) -> Chunk:
    """构造增强型 chunk 字典结构。"""
    meta = {
        "source": source_file.name,
        "chunk_seq": seq,
        "file_path": str(source_file),
        **(metadata or {})
    }
    chunk_id = f"{source_file.stem}#{seq}"
    return {
        "chunk_id": chunk_id,
        "text": text,
        "metadata": meta,
    }

def _md_header_splitter() -> MarkdownHeaderTextSplitter:
    """按工程认证文档常用的三级标题进行初步语义拆分。"""
    headers = [("#", "h1"), ("##", "h2"), ("###", "h3")]
    return MarkdownHeaderTextSplitter(headers_to_split_on=headers)

def _recursive_splitter(chunk_size: int = 800, chunk_overlap: int = 150) -> RecursiveCharacterTextSplitter:
    """递归字符分割器，优先保证中文句子的完整性。"""
    return RecursiveCharacterTextSplitter(
        chunk_size=chunk_size,
        chunk_overlap=chunk_overlap,
        separators=["\n\n", "\n", "。", "！", "？", "；", " ", ""],
    )

def _clean_table_whitespace(text: str) -> str:
    """
    专门针对 Markdown 表格单元格内的多余空格进行脱水。
    """
    lines = []
    for line in text.splitlines():
        # 仅对包含表格分隔符的行进行处理
        if '|' in line:
            # 1. 按照 | 分割
            parts = line.split('|')
            # 2. 对每个部分进行处理：
            #    - strip() 去除首尾空格
            #    - re.sub(r'\s+', ' ', ...) 将内部连续多个空格压缩为一个
            cleaned_parts = [re.sub(r'\s+', ' ', p.strip()) for p in parts]
            # 3. 重新组装，保留 Markdown 表格结构
            new_line = "|".join(cleaned_parts)
            lines.append(new_line)
        else:
            lines.append(line)
    return "\n".join(lines)

def _clean_text_density(text: str) -> str:
    """
    对 Markdown 文本进行“脱水”处理，提高信息密度。
    """
    # 1. 去除每行行首和行尾的空白字符
    text = _clean_table_whitespace(text)
    text = "\n".join([line.strip() for line in text.splitlines()])
    
    # 2. 压缩连续的换行符 (3个及以上压缩为2个，保留 Markdown 的段落感)
    text = re.sub(r'\n{3,}', '\n\n', text)
    
    # 3. 清理截图中的 <br> 标签 (可选，建议替换为空格或直接删除)
    text = text.replace("<br>", " ")
    
    # 4. 特殊处理：针对表格中间的大量空行
    # 如果一行只有 | 和空格，通常是空表格行，可以根据需求决定是否删除
    # text = re.sub(r'^\|[\s\|]+\|$', '', text, flags=re.MULTILINE)
    
    return text.strip()

def chunk_markdown_file(path: Path, chunk_size: int = 1300, chunk_overlap: int = 400) -> List[Chunk]:
    """
    处理带页码锚点的 Markdown 文件：
    1. 按标题层级做初步切分。
    2. 按字符长度做二次切分。
    3. 实时捕获并清理页码锚点，将页码存入 metadata。
    """
    raw_text = path.read_text(encoding="utf-8")
    
    text = _clean_text_density(raw_text)
    header_splitter = _md_header_splitter()
    length_splitter = _recursive_splitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)

    # 1. 语义分块（保留标题层级信息）
    docs = header_splitter.split_text(text)
    
    chunks: List[Chunk] = []
    seq = 1
    last_seen_page = 1  # 默认起始页

    for doc in docs:
        # 提取标题元数据
        headers = [str(v) for v in doc.metadata.values()]
        
        # 2. 长度分块
        sub_docs = length_splitter.split_documents([doc])
        
        for sub in sub_docs:
            content = sub.page_content
            
            # 3. 页码追踪：在分块中寻找注入的英文锚点
            # 匹配模式：==== PAGE_5_START ====
            page_matches = re.findall(r'==== PAGE_(\d+)_START ====', content)
            
            if page_matches:
                # 记录块内出现的所有页码（处理跨页情况）
                page_labels = sorted(list(set(int(p) for p in page_matches)))
                last_seen_page = page_labels[-1]  # 更新最后看到的页码
            else:
                # 若块内无锚点，则属于上个锚点的后续范围
                page_labels = [last_seen_page]

            # 4. 正文清理：剔除用于定位的锚点，保持知识库内容纯净
            # 防止大模型生成时带出 "==== PAGE_X_START ====" 噪声
            clean_text = re.sub(r'==== PAGE_\d+_(START|END) ====', '', content).strip()
            
            # 5. 组装 Chunk
            chunks.append(
                _build_chunk(
                    source_file=path,
                    text=clean_text,
                    seq=seq,
                    metadata={
                        "headers": headers,
                        "page_labels": page_labels,
                        "is_cross_page": len(page_labels) > 1
                    },
                )
            )
            seq += 1
            
    return chunks

def chunk_all_documents(
    md_root: Path | None = None,
    chunk_size: int = 1000,
    chunk_overlap: int = 250,
) -> List[Chunk]:
    """统一处理 data/parsed/md 下的所有已转换文件。"""
    

    all_chunks: List[Chunk] = []

    # 扫描目录下所有 .md 文件
    md_file = Path("大数据-2025级本科生培养方案-v13-724.md")
    print(md_file)
    all_chunks.extend(chunk_markdown_file(md_file, chunk_size=chunk_size, chunk_overlap=chunk_overlap))

    return all_chunks

if __name__ == "__main__":
    # 快速测试
    chunks = chunk_all_documents()
    print(chunks)
    print(f"✅ 处理完成，总分块数: {len(chunks)}")
```

```python
import json
import sys
from pathlib import Path
from typing import Iterable, List

def save_chunks_jsonl(chunks: Iterable[dict], output_path: Path) -> None:
    """将 chunk 列表写入 JSON Lines 文件，确认为 UTF-8 编码。"""
    output_path.parent.mkdir(parents=True, exist_ok=True)
    with output_path.open("w", encoding="utf-8") as f:
        for chunk in chunks:
            # ensure_ascii=False 保证中文不被转义，方便调试查看
            f.write(json.dumps(chunk, ensure_ascii=False))
            f.write("\n")
```
